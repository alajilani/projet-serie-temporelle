# 1- Deterministic component modeling

The deterministic component includes the trend and seasonal components.

### modeling only the trend component 

```r
# Determination of the descriptive variable related to the seasonal component

t<-c(1:length(time_series))
s.t<-(t-mean(t))/sd(t) # this variable is introduced in order to guarantee the same order of magnitude of the different descriptive variables
```

```r
Model0<-lm(time_series~s.t+I(s.t^2))
summary(Model0)
```

```r
plot(time_series, xlab="Time", ylab="Price",
     main="Brent Oil Prices  par jour",
     col="blue",lwd=2)
points(time(time_series),fitted(Model0),lwd=2,type="l",col="red")
legend("topleft", c("Original TS","Estimated TS"),inset = .02,
       bg = "gray90",lwd=c(2,2),col=c("blue","red"))
```

```r
acf(Model0$residuals)
```

This improved model includes both a linear trend and a quadratic term, which allows it to better capture the nonlinear shape of the time series. The non-significant coefficients and the relatively low R² value (0.5316) indicate a very poor fit to the data, suggesting that there may still be unmodeled patterns, which implies that a more thorough exploration of the temporal structures could be beneficial to obtain a better modeling of the series.

### Modeling the trend and seasonal components without break points


```r
T=365
MC=matrix(0,length(time_series),(T/2))
MS=matrix(0,length(time_series),(T/2))
for(i in 1:(T/2)){MC[,i]=cos(2*pi*i*t/T)}
for(i in 1:(T/2)){MS[,i]=sin(2*pi*i*t/T)}
```

```r
Model1<-lm(time_series~s.t+I(s.t^2)+I(t^3)+MS+MC)
summary(Model1)
```

```r
plot(time_series, xlab="Time", ylab="Price",
     main="Brent Oil Prices  par jour",
     col="blue",lwd=2)
points(time(time_series),fitted(Model1),lwd=2,type="l",col="red")
legend("topleft", c("Original TS","Estimated TS"),inset = .02,
       bg = "gray90",lwd=c(2,2),col=c("blue","red"))
```

The model captures a significant portion of the variability in crude oil prices and is statistically significant. However, the smoother red curve shows that the model captures the overall trend well but doesn't fully capture the short-term fluctuations in actual crude oil prices. This suggests that additional modeling approaches may be needed to better represent short-term variations.

```r
# with dummy variables
# n=number of periods, k= index, T=period
VD<-function(n,k,T)
{x<-rep(0,T*n)
for (i in 0:(n-1)){x[k+i*T]<-1}
x
}
MVD<-matrix(0,12970,10)
for(i in 1:10) MVD[,i]<-VD(1297.0,i,10)
MVDF<-MVD[1:(1296.3*10),]

MVDF
```

```r
Model2<-lm(time_series~s.t+I(s.t^2)+I(s.t^3)+MVDF)
summary(Model2)
```

```r
plot(time_series, xlab="Time", ylab="Price",
     main="Brent Oil Prices  par jour",
     col="blue",lwd=2)
points(time(time_series),fitted(Model2),lwd=2,type="l",col="red")
legend("topleft", c("Original TS","Estimated TS"),inset = .02,
       bg = "gray90",lwd=c(2,2),col=c("blue","red"))
```

## modeling with breakpoints


```r
# Installer et charger le package strucchange

library(strucchange)

# Créer un vecteur de temps
t <- 1:length(time_series)

# Utiliser breakpoints() pour détecter les points de rupture
bp <- breakpoints(time_series ~ t)
summary(bp)

bp$breakpoints
```

```r
plot(time_series, ylab = "Price")
lines(bp)
```

```r
dummy11<-ifelse(t>bp$breakpoints[1]& t<=bp$breakpoints[2],1,0)
dummy21<-ifelse(t>bp$breakpoints[2] & t<=bp$breakpoints[3],1,0)
dummy31<-ifelse(t>bp$breakpoints[3] & t<=bp$breakpoints[4],1,0)
dummy41<-ifelse(t>bp$breakpoints[4] & t<=bp$breakpoints[5],1,0)
dummy51<-ifelse(t>bp$breakpoints[5],1,0)
```

```r
Model3<-lm(time_series~t*dummy11+t*dummy21+t*dummy31+t*dummy41+t*dummy51+MVDF)
summary(Model3)
```

```r
plot(time_series, xlab="Time", ylab="Price",
     main="Brent Oil Prices  par jour",
     col="blue",lwd=2)
points(time(time_series),fitted(Model3),lwd=2,type="l",col="red")
legend("topleft", c("Original TS","Estimated TS"),inset = .02,
       bg = "gray90",lwd=c(2,2),col=c("blue","red"))
```

```r
Model4<-lm(time_series~t*dummy11+t*dummy21+t*dummy31+t*dummy41+t*dummy51+MC+MS)
summary(Model4)
```

```r
plot(time_series, xlab="Time", ylab="Price",
     main="Brent Oil Prices  par jour",
     col="blue",lwd=2)
points(time(time_series),fitted(Model4),lwd=2,type="l",col="red")
legend("topleft", c("Original TS","Estimated TS"),inset = .02,
       bg = "gray90",lwd=c(2,2),col=c("blue","red"))
```

The introduction of the breakpoints into the model has indeed significantly improved the model's ability to faithfully represent the complex dynamics of this time series, in comparison to the model without these structural breaks. This indicates that this approach has been beneficial

VII.Modeling the Stochastic (Residual) Component</p>

### Graphical Representation of the Residual Component

```r
sp.res<-Model9$residuals
plot(time(time_series),sp.res, xlab="Time", ylab="",
     main="Residual Component",
     col="purple4",lwd=2,type="l")
```

### The curve is sinusoidal, so there is no white noise

```r
acf(sp.res)
```

### According to the ACF, there is a threshold exceedance, indicating that the noise is not white.

## Stationarity Test

```r
library(tseries)
adf.test(sp.res)
```

### The p-value= 0.01 < 0.05 => therefore, we accept the alternative hypothesis: the data is stationary.

## White Noise Test

```r
Box.test(sp.res)
     
```

### The p-value = p-value < 2.2e-16 < 0.05 indicates that the alternative hypothesis is favored: the stationary process does not behave like white noise.

## ARMA Model for the Residual Component

```r
par(mfrow=c(1,2))
acf(sp.res,120)
pacf(sp.res,120)
```

#### We will use the auto.arima() function from the forecast package to determine the order 'p' of the AR model and the order 'q' of the MA model, as it is difficult to determine them solely from the ACF and PACF.

```r
library(forecast)
#ARIMA.d=auto.arima(d.sp.res,trace=TRUE)
ARIMA=auto.arima(sp.res,d=0,trace=TRUE)
```

```r
Model.res<-arima(sp.res,order = c(1, 0, 1))
```

```r

adf.test(Model.res$residuals)
```


####  p-value = 0.01 < 0.05, Therefore, Model.res$residuals are stationary

```r
checkresiduals(Model.res)
```

```r
library(aTSA)
ts.diag(Model.res)
```

```r
Box.test(Model.res$residuals)
```


#### The p-value  = 0.9797 > 0.05, indicating that the alternative hypothesis is favored: the stationary process behaves like white noise. (So, Model.res is a good model)

```r
# Générer les prévisions
forecast_values <- forecast(Model.res,150)
```

# SARIMA Model


```r

Model.SARIMA<-auto.arima(sp.res)
summary(Model.SARIMA)
```

```r

Model.SARIMA.fit<-arima(sp.res, order = c(1, 0, 1),seasonal = list(order = c(0,0,0), period = 365))
checkresiduals(Model.SARIMA.fit)
```

```r

Box.test(Model.SARIMA.fit$residuals)
```


#### ===> The Box-Pierce test gives a p-value for our new SARIMA model that is greater than 0.05, indicating that it is a good model.
